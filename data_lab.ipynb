The `class_float` column is also a bit hard to read, because the classes are represented by numbers. This can be preferable for some algorithms, but for our use case today we will replace these values by explicit names. For this, we will use the `DataFrame.apply()` function.

```python
# This is the function we will apply to the "class_float" column
def name_mapping(number:float):
    """This function maps 0.0, 1.0 and 2.0 to their corresponding values in the iris dataset."""

    name_map = {
        0.0: "setosa", # replace by the correct name
        1.0: "versicolor", # replace by the correct name
        2.0: "virginica", # replace by the correct name
    }

    if number not in name_map.keys(): # making sure the number is one of the expected values
        raise ValueError("Not a valid number!")

    return name_map[number] # This is an alternative to using a lot of if/else blocks
```

```python
# We can now apply the function
df["species"] = df["species"].apply(name_mapping) # We pass the function as parameter, not its result! This is why we must not use parentheses.
```

```python
cols = ['worst concave points','worst perimeter', 'mean concave points', 'worst radius', 'target'  ]
key_features = df[cols]
```

1-worst concave points
Measures the largest number of concave portions on the tumor’s contour in the worst (most extreme) image sample.

A higher value often indicates a more irregular, likely malignant tumor.

This is one of the most discriminative features.


2️-worst perimeter

Measures the largest perimeter of the tumor across all images/slices.

Tumors with higher perimeters tend to be malignant, because cancerous tumors are often larger and irregularly shaped.

3-mean concave points

The average number of concave points across all tumor measurements.

Concave points reflect indentations in the tumor’s shape.

Higher average concavity usually correlates with malignancy.

4-worst radius

The largest radius of the tumor observed in the images.

Malignant tumors often have larger radii than benign ones.

Note: radius, perimeter, area, concave points are all morphological features that describe size and shape irregularity.
